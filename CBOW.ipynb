```python
data = """We are about to study the idea of a computational process.
Computational processes are abstract beings that inhabit computers.
As they evolve, processes manipulate other abstract things called data.
The evolution of a process is directed by a pattern of rules
called a program. People create programs to direct processes. In effect,
we conjure the spirits of the computer with our spells."""
```

Data Preprocessing

```python
data=data.lower()
print(data)
```

Remove Punctuation

```python
import string
data=data.translate(str.maketrans('', '', string.punctuation))
print(data)
```

```python
 words=data.split()
```

```python
print(words)
```

```python
!python -m spacy download en_core_web_lg
import spacy
nlp = spacy.load('en_core_web_lg')
```

Split Data

```python
data=[]
for i in range(1,len(words)):
  if i%3==0:
    data.append([words[i-3],words[i-2],words[i-1]])

print(data)
```

Split Data Into Target And Data

```python
X=[]
y=[]
for i in range(len(data)):
    X.append([data[i][0],data[i][1]])
    y.append([data[i][2]])
print(X)
print(y)
```

Emembeddings

```python
X_vector=[]
y_vector=[]
for x in range(len(X)):
   X_vector.append([nlp.vocab[X[x][0]].vector,nlp.vocab[X[x][1]].vector])
for y_val in range(len(y)):
    y_vector.append(nlp.vocab[y[y_val][0]].vector)
```
